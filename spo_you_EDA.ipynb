## Outlier Values


```python
plt.figure(figsize=(14, 6))

# Views box-plot
plt.subplot(1, 3, 1)
sns.boxplot(y=df['Views'])
plt.title('Views')

# Likes  box-plot
plt.subplot(1, 3, 2)
sns.boxplot(y=df['Likes'])
plt.title('Likes')

# Comments  box-plot
plt.subplot(1, 3, 3)
sns.boxplot(y=df['Comments'])
plt.title('Comments')

plt.tight_layout()
plt.show()
```

```python
# Outlier detection for Views ; 

Q1 = df['Views'].quantile(0.25)
Q3 = df['Views'].quantile(0.75)
IQR1 = Q3 - Q1

print(IQR1)
```

```python
outlier_rows1 = df[(df['Views'] < (Q1 - 2 * IQR1)) | (df['Views'] > (Q3 + 2 * IQR1))].index
print(outlier_rows1)
```

```python
df = df[~((df['Views'] < (Q1 - 2 * IQR1)) | (df['Views'] > (Q3 + 2 * IQR1)))]
```

```python
# Outlier detection for Likes ; 

Q1 = df['Likes'].quantile(0.25)
Q3 = df['Likes'].quantile(0.75)
IQR2 = Q3 - Q1

print(IQR2)
```

```python
outlier_rows2 = df[(df['Likes'] < (Q1 - 2 * IQR2)) | (df['Likes'] > (Q3 + 2 * IQR2))].index
print(outlier_rows2)
```

```python
df = df[~((df['Likes'] < (Q1 - 2 * IQR2)) | (df['Likes'] > (Q3 + 2 * IQR2)))]
```

```python
# Outlier detection for Comments ; 

Q1 = df['Comments'].quantile(0.25)
Q3 = df['Comments'].quantile(0.75)
IQR3 = Q3 - Q1

print(IQR3)
```

```python
outlier_rows3 = df[(df['Comments'] < (Q1 - 2 * IQR3)) | (df['Comments'] > (Q3 + 2 * IQR3))].index
print(outlier_rows3)
```

```python
df = df[~((df['Comments'] < (Q1 - 2 * IQR3)) | (df['Comments'] > (Q3 + 2 * IQR3)))]
```

```python
plt.figure(figsize=(14, 6))

# Views 
plt.subplot(1, 3, 1)
sns.boxplot(y=df['Views'])
plt.title('Views')

# Likes 
plt.subplot(1, 3, 2)
sns.boxplot(y=df['Likes'])
plt.title('Likes')

# Comments 
plt.subplot(1, 3, 3)
sns.boxplot(y=df['Comments'])
plt.title('Comments')

plt.tight_layout()
plt.show()
```

# <p style="background-color:turquoise; font-family:newtimeroman; font-size:150%; text-align:center; border-radius: 15px 50px;"> Comparison of most-streamed Spotify and YouTube songs</p>


```python
data = df.groupby("Artist").agg({"Stream" : "sum", 
                                 "Views" : "sum"}).sort_values(by = [
                           "Stream", "Views"], ascending = [False, False]).head(30).reset_index()
data
```

```python
# Reshaping data
data_melted = data.melt(id_vars='Artist', 
                        value_vars=['Stream', 'Views'], 
                        var_name='Type', 
                        value_name='Value')

# Visualization
plt.figure(figsize=(10,10))

sns.barplot(x='Value', 
            y='Artist', 
            hue='Type', 
            data=data_melted)

plt.title('Total Spotify Streams and YouTube Views by Artist')
plt.xlabel('Total Value')
plt.ylabel('Artist')
plt.xticks(rotation=90)

sns.despine()

plt.show()
```

### As you can see, there are a lot of differences between Spotify stream and YouTube views. We can understand that, people use Spotify for play music much more than YouTube

Our data shows that music streamed on Spotify has a higher rate of being streamed on YouTube. This is a reflection of how people perceive music across platforms. Spotify is a platform that focuses more on music, while YouTube appeals to a broader audience. This may also be supported by the fact that music labels promote more on Spotify and less on YouTube.

# <p style="background-color:turquoise; font-family:newtimeroman; font-size:150%; text-align:center; border-radius: 15px 50px;"> Channel Statistics</p>


```python
channel_data = df[['Channel', 'Views', 'Likes', 'Comments']]

channel_data = channel_data.groupby('Channel').agg({'Views': 'mean', 
                                                          'Likes': 'mean', 
                                                          'Comments': 'mean'}).sort_values(by = [
                          "Views", "Likes", "Comments"], ascending = [False, False, False]).head(5).reset_index()
channel_data
```

```python
plt.figure(figsize=(18,8))


bar_colors = ['#13678A', '#45C4B0', '#9AEBA3']

# Views
plt.subplot(1, 3, 1)
plt.bar(channel_data['Channel'], 
        channel_data['Views'], 
        color=bar_colors[0])
plt.xlabel('Channel')
plt.ylabel('Total Views')
plt.title('Total Views')
plt.xticks(rotation=90, fontsize=15)

# Likes
plt.subplot(1, 3, 2)
plt.bar(channel_data['Channel'], 
        channel_data['Likes'],
        color=bar_colors[1])
plt.xlabel('Channel')
plt.ylabel('Total Likes')
plt.title('Total Likes')
plt.xticks(rotation=90, fontsize=15)

# Comments

plt.subplot(1, 3, 3)
plt.bar(channel_data['Channel'], 
        channel_data['Comments'], 
        color=bar_colors[2])
plt.xlabel('Channel')
plt.ylabel('Total Comments')
plt.title('Total Comments')
plt.xticks(rotation=90, fontsize=15)

plt.tight_layout()
plt.show()
```
